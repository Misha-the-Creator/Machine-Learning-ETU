```python
import pandas as pd
```

```python
# 2.Создать Python скрипт. Загрузить данные в датафрейм

all_data = pd.read_csv('groceries - groceries.csv') 
all_data
```

```python
# 3. Переформируем данные, удалив все значения NaN

np_data = all_data.to_numpy() 
np_data = [[elem for elem in row[1:] if isinstance(elem,str)] for row in np_data]
```

```python
# 4. Получим список всех уникальных товаров
unique_items = set() 

for row in np_data: 
    for elem in row: 
        unique_items.add(elem)
```

```python
# 5. Выведите список товаров, а также их количество

print(f"Кол-во товаров: {len(unique_items)}")
print('---------------------------------------')
print(f"Список товаров: {unique_items}")
```

```python
# 1. Преобразуем данные к виду, удобному для анализа

from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder() 
te_ary = te.fit(np_data).transform(np_data) 
data = pd.DataFrame(te_ary, columns=te.columns_)
data
```

```python
# 2. Проведем ассоциативный анализ используя алгоритм FPGrowth при уровне поддержки 0.03

from mlxtend.frequent_patterns import fpgrowth 

result_fpgrowth = fpgrowth(data, min_support=0.03, use_colnames=True)
result_fpgrowth['length'] = result_fpgrowth['itemsets'].apply(lambda x: len(x))
result_fpgrowth
```

```python
# 3. Проанализируйте получившиеся варианты. 
# Определите минимальное и максимальное значения для уровня поддержки для набора из 1,2, и.т.д. объектов.

result_fpgrowth.groupby('length').agg({'support': ['min', 'max']})
```

```python
# 4. Проведите аналогичный анализ используя алгоритм FPMax

from mlxtend.frequent_patterns import fpmax
result_fpmax = fpmax(data, min_support=0.03)
result_fpmax['length'] = result_fpmax['itemsets'].apply(lambda x: len(x))
result_fpmax
```

```python
result_fpmax.groupby('length').agg({'support': ['min', 'max']})
```

```python
# 6. Постройте гистограмму для каждого товара. 
# Столбцы на гистограмме должны быть упорядочены по уменьшению частоты. 
# Отобразите результат только для 10 самых встречаемых товаров. 
# Как данная гистограмма коррелирует с результатами?

import matplotlib.pyplot as plt
from collections import Counter

result_fpgrowth_sorted = result_fpgrowth.query("length == 1").sort_values(by='support', ascending=False).head(10)

item_support = Counter()
for support, itemset in zip(result_fpgrowth_sorted['support'], result_fpgrowth_sorted['itemsets']):
    for item in itemset:
        item_support[item] += support

support_df = pd.DataFrame(item_support.items(), columns=['Товар', 'Поддержка'])

# Построение гистограммы
plt.figure(figsize=(10, 6))
plt.bar(support_df['Товар'], support_df['Поддержка'])
plt.xlabel('Товары')
plt.ylabel('Частота')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Преобразуем набор данных, чтобы он содержал ограниченный набор товаров

items = ['whole milk', 'yogurt', 'soda', 'tropical fruit', 'shopping bags',
         'sausage', 'whipped/sour cream', 'rolls/buns', 'other vegetables',
         'root vegetables', 'pork', 'bottled water', 'pastry', 'citrus fruit',
         'canned beer', 'bottled beer'] 

np_data = all_data.to_numpy() 
np_data = [[elem for elem in row[1:] if isinstance(elem,str) and elem in items] for row in np_data]
```

```python
#8. Проведите анализ FPGrowth и FPMax для нового набора данных. Проанализируйте, что изменилось

te = TransactionEncoder() 
te_ary = te.fit(np_data).transform(np_data) 
data = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
result_fpgrowth = fpgrowth(data, min_support=0.03, use_colnames=True)
result_fpgrowth['length'] = result_fpgrowth['itemsets'].apply(lambda x: len(x))
result_fpgrowth
```

```python
result_fpmax = fpmax(data, min_support=0.03)
result_fpmax['length'] = result_fpmax['itemsets'].apply(lambda x: len(x))
result_fpmax
```

```python
# 9. Постройте график изменения количества получаемых правил от уровня поддержки.
# На графике отдельно отобразите кривые для набора товаров 1, 2, и.т.д.
# Какие выводы можно сделать по данному графику?

def sss(sup):
    result_fpgrowth = fpgrowth(data, min_support=sup, use_colnames=True)
    result_fpgrowth['length'] = result_fpgrowth['itemsets'].apply(lambda x: len(x))
    return len(result_fpgrowth)

support=[]
length=[]

for p in range(1, 20, 1):
    support.append(p/100)
    length.append(sss(p/100))

print(support)
print(length)
```

```python
graph = pd.DataFrame({'support': support, 'length': length})

plt.plot(graph.support, graph.length)
plt.xlabel('Уровень поддержки')
plt.ylabel('Количество правил')
plt.show()
```

```python
# Ассоциативные правила
# 1. Сформируем набор данных из определенных товаров и так, чтобы размер транзакции был 2 и более

np_data = all_data.to_numpy() 
np_data = [[elem for elem in row[1:] if isinstance(elem,str) and elem in items] for row in np_data] 
np_data = [row for row in np_data if len(row) > 1]
```

```python
# 2. Получим частоты наборов используя алгоритм FPGrowth

te = TransactionEncoder()
te_ary = te.fit(np_data).transform(np_data) 
data = pd.DataFrame(te_ary, columns=te.columns_)

result = fpgrowth(data, min_support=0.05, use_colnames = True)
result.head()
```

```python
# 3. Проведем ассоциативный анализ 

from mlxtend.frequent_patterns import association_rules
rules = association_rules(result, min_threshold = 0.3) 
rules.head()
```

```python
rules = association_rules(result, min_threshold = 0.4, metric='confidence')
rules
```

```python
import networkx as nx
```

```python
G = nx.DiGraph()

# Добавляем вершины и рёбра в граф
for index, row in rules.iterrows():
    antecedent = frozenset(row['antecedents'])  # Используем frozenset для антецедента
    consequent = frozenset(row['consequents'])  # Используем frozenset для консеквента
    support = row['support']
    confidence = row['confidence']
    
    # Добавляем ребро, ориентированное от антецедента к консеквенту
    G.add_edge(antecedent, consequent, weight=support, confidence=confidence)

# Создаем человекочитаемые подписи вершин
labels = {node: ', '.join(map(str, node)) for node in G.nodes()}

# Нормализуем веса для визуализации
widths = [data['weight'] * 10 for u, v, data in G.edges(data=True)]  # Увеличим ширину рёбер
edge_labels = {(u, v): f"{data['confidence']:.2f}" for u, v, data in G.edges(data=True)}  # Подписи на рёбрах

# Рисуем граф
pos = nx.spring_layout(G, k=0.5)  # Положение вершин
nx.draw(G, pos, labels=labels, with_labels=True, node_color='lightblue', node_size=3000, font_size=10, font_color='black', arrows=True)

# Рисуем рёбра с соответствующей шириной и подписями
nx.draw_networkx_edges(G, pos, width=widths, alpha=0.5)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Граф ассоциативных правил")
plt.show()
```
